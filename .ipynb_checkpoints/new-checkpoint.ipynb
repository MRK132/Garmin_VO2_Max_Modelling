```python
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression 
from pyearth import Earth 
import dask.dataframe as dd
import dask.bag as db
from tqdm import tqdm
```

```python
#a = dd.read_parquet("2020_H1/20210108_114301_00074_5nk2u_7b1fb6ad-aadf-4da3-8710-8aa43440b84e.snappy.parquet")
a = dd.read_parquet("2020_H1/*.snappy.parquet")
```

```python
af = a.compute()
#af = af.reset_index()
#af['tru_ind'] = af.index

# high_throughput = []
# for i in a['_energy_throughput_battery_kwh']:
#     if i > 28:
#         high_throughput.append(i)
#     else:
#         high_throughput.append(None)
        

fast_charges = []
for i in a['_energy_throughput_battery_kwh']:
    if i > 60:
        fast_charges.append(i)
    else:
        fast_charges.append(None)
        
ambient_high = []
ambient_low = []
ambient_med = []
ambient_level = []
for i in a['cycle_avg_ambienttemp']:
    if i > 20:
        ambient_high.append(i)
    else:
        ambient_high.append(None)
        
for i in a['cycle_avg_ambienttemp']:      
    if i < 3:
        ambient_low.append(i)
    else:
        ambient_low.append(None)
        
for i in a['cycle_avg_ambienttemp']:      
    if 3 <= i < 20:
        ambient_med.append(i)
    else:
        ambient_med.append(None)

#af = a.compute()


for i in a['cycle_avg_ambienttemp']:
    if i > 20:
        ambient_level.append(3)
    
        
        
    elif i < 3:
        ambient_level.append(1)
    else:
        ambient_level.append(2)



#af['high_throughput'] = high_throughput
af['fast_charges'] = fast_charges
af['ambient_low'] = ambient_low
af['ambient_med'] = ambient_med
af['ambient_high'] = ambient_high
af['ambient_level'] = ambient_level

#del high_throughput
del fast_charges
del ambient_low
del ambient_med
del ambient_high
del ambient_level

af['cycle_ambient_temp_fluctuation'] = af['cycle_max_ambienttemp'] - af['cycle_min_ambienttemp']

high_fluc = []
for i in af['cycle_ambient_temp_fluctuation']:
        if i >= 10:
            high_fluc.append(1)
        else:
            high_fluc.append(0)

af['high_fluc']=high_fluc
del high_fluc


af['energy_flow_through_battery_kwh'] = af['_energy_into_battery_kwh'] + af['_energy_outof_battery_kwh']

af['soc_change_cycle'] = af['_soc_end_cycle'] - af['_soc_start_cycle']

soc_level_end = []
for i in af['_soc_end_cycle']: # 
        if i > 85:
            soc_level_end.append(3)
        elif i < 30:
            soc_level_end.append(1)
        else:
            soc_level_end.append(2)

af['soc_state_end_of_cycle']=soc_level_end
del soc_level_end

soc_level_start = []
for i in af['_soc_start_cycle']: # 
        if i > 85:
            soc_level_start.append(3)
        elif i < 30:
            soc_level_start.append(1)
        else:
            soc_level_start.append(2)

af['soc_state_start_of_cycle']=soc_level_start
del soc_level_start


af['cycle_fluc_hvbattcelltempcoolest_'] = af['cycle_max_hvbattcelltempcoolest_'] - af['cycle_min_hvbattcelltempcoolest_']

af['cycle_fluc_hvbattfusetemperature'] = af['cycle_max_hvbattfusetemperature'] - af['cycle_min_hvbattfusetemperature']

af['cycle_fluc_hvbattmemtemperature'] = af['cycle_max_hvbattmemtemperature'] - af['cycle_min_hvbattmemtemperature']

af['cycle_fluc_hvbattinletcoolanttemp'] = af['cycle_max_hvbattinletcoolanttemp'] - af['cycle_min_hvbattinletcoolanttemp']




#####################################################################################################
af = af.sort_values(by=['rehashvin', '_end_time_cycle'])
af = af.reset_index()
af['tru_ind'] = af.index
```

```python
for row in tqdm(range(len_)):
    
    km_driven = af["distance_driven_km_int"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    km_driven_for_car_till_this_soh_reading = sum(km_driven)
    
    time_driven = af["cycle_duration_min"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    time_driven_for_car_till_this_soh_reading = sum(time_driven)

    soh_present['km_driven_per_soh_reading_per_vin'].iloc[row] =  km_driven_for_car_till_this_soh_reading
    soh_present['time_driven_per_soh_reading_per_vin'].iloc[row] = time_driven_for_car_till_this_soh_reading
```

```python
for row in tqdm(range(len_)):
    

    temp_af1 = af['_cycle'].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    temp_af2 = af[ 'soc_state_start_of_cycle'].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    temp_af3 = af[ 'distance_driven_km_int'].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    temp_af4 = af[ 'cycle_duration_min' ].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    temp_af5 = af['ambient_level'].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    temp_af = pd.DataFrame({'_cycle' : temp_af1,'soc_state_start_of_cycle' : temp_af2,
                            'distance_driven_km_int': temp_af3, 'cycle_duration_min': temp_af4,
                           'ambient_level' : temp_af5})      

     
    
    
    temp_af_low_state_dr = temp_af[(temp_af['_cycle'] == 'drive') & (temp_af['soc_state_start_of_cycle']==1)]
    km_driven_in_low_soc_state = sum(temp_af_low_state_dr['distance_driven_km_int'])
    time_driven_in_low_soc_state = sum(temp_af_low_state_dr['cycle_duration_min'])
    del temp_af_low_state_dr
    
    temp_df_high_temp_dr = temp_af[(temp_af['_cycle'] == 'drive') & (temp_af['ambient_level']==3)]
    km_driven_in_high_temperature = sum(temp_df_high_temp_dr['distance_driven_km_int'])
    time_driven_in_high_temperature = sum(temp_df_high_temp_dr['cycle_duration_min'])
    del temp_df_high_temp_dr
    
    temp_df_low_temp_dr = temp_af[(temp_af['_cycle'] == 'drive') & (temp_af['ambient_level']==1)]
    km_driven_in_low_temperature = sum(temp_df_low_temp_dr['distance_driven_km_int'])
    time_driven_in_low_temperature = sum(temp_df_low_temp_dr['cycle_duration_min'])
    del temp_df_low_temp_dr

    
    
    
    soh_present['km_driven_in_low_soc_state'].iloc[row] = km_driven_in_low_soc_state
    soh_present['time_driven_in_low_soc_state'].iloc[row] = time_driven_in_low_soc_state
    soh_present['km_driven_in_low_temperature'].iloc[row] = km_driven_in_low_temperature
    soh_present['time_driven_in_low_temperature'].iloc[row] = time_driven_in_low_temperature
    soh_present['km_driven_in_high_temperature'].iloc[row] = km_driven_in_high_temperature
    soh_present['time_driven_in_high_temperature'].iloc[row] = time_driven_in_high_temperature
```

```python
for row in tqdm(range(len_)):
    fast_charges_per_soh_reading = af["fast_charges"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    fast_charges_count = fast_charges_per_soh_reading.dropna()
    
    ambient_temp_fluctuations = af["cycle_ambient_temp_fluctuation"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    ambient_temp_fluctuations_till_soh = sum(ambient_temp_fluctuations)
    
    energy_flow_through_battery_ = af["energy_flow_through_battery_kwh"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    energy_flow_through_battery_soh = sum(energy_flow_through_battery_)

    soh_present['fast_charges_per_soh_reading'].iloc[row] = len(fast_charges_count)
    soh_present['ambient_temp_fluctuations_per_soh'].iloc[row] = ambient_temp_fluctuations_till_soh
    soh_present['energy_flow_through_battery_soh'].iloc[row] = energy_flow_through_battery_soh
```

```python
for row in tqdm(range(len_)):
    cycle_fluc_hvbattcelltempcoolest_ = af["cycle_fluc_hvbattcelltempcoolest_"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    soh_fluc_hvbattcelltempcoolest_ = sum(cycle_fluc_hvbattcelltempcoolest_)
    
    cycle_fluc_hvbattfusetemperature = af["cycle_fluc_hvbattfusetemperature"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    soh_fluc_hvbattfusetemperature = sum(cycle_fluc_hvbattfusetemperature)

    cycle_fluc_hvbattmemtemperature = af["cycle_fluc_hvbattmemtemperature"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    soh_fluc_hvbattmemtemperature = sum(cycle_fluc_hvbattmemtemperature)

    cycle_fluc_hvbattinletcoolanttemp = af["cycle_fluc_hvbattinletcoolanttemp"].iloc[soh_present.iloc[row,-2]:soh_present.iloc[row,-1]]
    soh_fluc_hvbattinletcoolanttemp = sum(cycle_fluc_hvbattinletcoolanttemp)
    
    soh_present['soh_fluc_hvbattcelltempcoolest_'].iloc[row] = soh_fluc_hvbattcelltempcoolest_
    soh_present['soh_fluc_hvbattfusetemperature'].iloc[row] = soh_fluc_hvbattfusetemperature
    soh_present['soh_fluc_hvbattmemtemperature'].iloc[row] = soh_fluc_hvbattmemtemperature
    soh_present['soh_fluc_hvbattinletcoolanttemp'].iloc[row] = soh_fluc_hvbattinletcoolanttemp
```

```python
#from xgboost.sklearn import XGBRegressor
import matplotlib.pyplot as plt
import shap
```

```python
from dask.distributed import Client
#client = Client('scheduler-address:8786')
client = Client()

from dask_ml.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#from xgboost.sklearn import XGBRegressor
from xgboost import XGBRegressor
model = XGBRegressor()


from sklearn.model_selection import RandomizedSearchCV
import numpy as np
# Hyperparameter values to test
param_grid = {'max_features': ['auto', 'sqrt'],
              'max_depth': np.arange(5, 36, 5),
              'min_samples_split': [5, 10, 20, 40],
              'min_samples_leaf': [2, 6, 12, 24],
             }
# RandomizedSearchCV with a limit of 10 combinations to test
#model = RandomizedSearchCV(estimator = XGBRegressor(), param_distributions = param_grid, n_iter = 10, verbose = 2, n_jobs = -1)
#tree_reg.fit(X_train, y_train)


import joblib
from sklearn.externals.joblib import parallel_backend
with parallel_backend('dask'):
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
```
